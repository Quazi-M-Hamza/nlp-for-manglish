```python
from selenium import webdriver
from tqdm import tqdm
# Importing the ChromeDriver class for handling webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup
import re
import path
import os
```

```python
def is_english(s: str) -> bool:
    if len(re.findall(u'[\u0D00-\u0D7F]', s)) <= 100:
        return True
    return False
```

```python
source_folder = 'valid_ma'
all_files = os.listdir(source_folder)
print(len(all_files))
```

```python
dest_folder = 'valid_en'
done_files = os.listdir(dest_folder)
print(len(done_files))
```

```python
base_url = 'file:///home/hap-101/Downloads/fire/code-mixed-enma/malyalam_wiki_articles'
for filename in tqdm(all_files):
    try:
        if filename in done_files:
            continue
        url = f'{base_url}/{source_folder}/{filename}'
        sleep_time = 4
        trial = 1
        in_english = False
        while trial <= 1 and not in_english:
            options = webdriver.ChromeOptions()
            prefs = {
              "translate_whitelists": {"ml":"en"},
              "translate":{"enabled":"true"}
            }
            options.add_experimental_option("prefs", prefs)
            browser = webdriver.Chrome(ChromeDriverManager().install(), chrome_options=options)
            browser.get(url)
            time.sleep(sleep_time)
            alltext = browser.page_source
            soup = BeautifulSoup(alltext, 'html.parser')
            alltext = soup.find('pre').text
            in_english = is_english(alltext)
            if in_english:
                with open(f'{dest_folder}/{filename}', 'w') as f:
                    f.write(alltext)
            trial += 1
            sleep_time += 4
            time.sleep(3)
            browser.quit()
        if not in_english:
            print(f'Failed for {url}')
    except:
        continue
```
